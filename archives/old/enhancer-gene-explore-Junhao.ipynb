## Correlation  

```python
_tmp = enh_atac_clsts.reset_index()
enh_lists = _tmp['chr'] + '_' + _tmp['start'].astype(str) + '_' + _tmp['end'].astype(str)
```

```python
# full correlation

# gene 
_x_features = gene_rna_clsts.index.values
_X = gene_rna_clsts.values

# # enhancer
_y_features = enh_lists.values
_Y = enh_atac_clsts.values 
print(_X.shape, _Y.shape,)
```

```python
_X_ranks = snmcseq_utils.rank_rows(_X)
_Y_ranks = snmcseq_utils.rank_rows(_Y)
```

```python
ti = time.time()
# val, dist
_x_idx = snmcseq_utils.get_index_from_array(_x_features, to_evals['gene'].values)
_y_idx = snmcseq_utils.get_index_from_array(_y_features, to_evals['ens'].values)

to_evals['val'] = snmcseq_utils.vcorrcoef(_X_ranks[_x_idx], _Y_ranks[_y_idx]) 

print(time.time()-ti)
to_evals.head()
output = '/cndd2/fangming/test/for_junhao/gene_enh_corr.tsv'
to_evals.to_csv(output, sep='\t', header=True, index=False)
```

```python
ti = time.time()
to_evals_shuffled = to_evals.copy()
# val, dist
_x_features_shuffled = np.random.permutation(_x_features)
_y_features_shuffled = np.random.permutation(_y_features)

_x_idx = snmcseq_utils.get_index_from_array(_x_features_shuffled, to_evals_shuffled['gene'].values)
_y_idx = snmcseq_utils.get_index_from_array(_y_features_shuffled, to_evals_shuffled['ens'].values)

to_evals_shuffled['val'] = snmcseq_utils.vcorrcoef(_X_ranks[_x_idx], _Y_ranks[_y_idx]) 

print(time.time()-ti)
to_evals_shuffled.head()
output = '/cndd2/fangming/test/for_junhao/gene_enh_corr_shuffled.tsv'
to_evals_shuffled.to_csv(output, sep='\t', header=True, index=False)
```

```python
sns.distplot(to_evals['val'])
sns.distplot(to_evals_shuffled['val'])
```

```python
fig, ax = plt.subplots() 
out1 = snmcseq_utils.gen_cdf(to_evals['val'].values, ax, flip=True)
out2 = snmcseq_utils.gen_cdf(to_evals_shuffled['val'].values, ax, flip=True)
ax.set_xlabel('corr')
ax.set_ylabel('cumulative fraction of pairs')
plt.show()
```

```python
x, y = out1
x2, y2 = out2

xbin = np.linspace(0, 1, 1000)
fdr = np.interp(xbin, x2, y2)/np.interp(xbin, x, y)

fig, ax = plt.subplots()
ax.plot(xbin, fdr)
ax.set_xlabel('corr')
ax.set_ylabel('FDR')
plt.show()
```
