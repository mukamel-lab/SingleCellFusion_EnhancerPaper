# get examples for all possible genes

```python
example_genes = look_evals.gene.unique()
coord_by_gene = {}
for i in example_genes:
    use_enhs = look_evals.loc[look_evals.gene == i].ens.values
    enh_df = enhs.loc[enhs[3].isin(use_enhs)]
    gene_df = genes.loc[genes[3]==i]

    use_df = pd.concat([enh_df, gene_df])
    
    use_df.columns = [ 'chrom', 'start', 'end', 'annot']
    
    coord_by_gene[i] = use_df
    
```
