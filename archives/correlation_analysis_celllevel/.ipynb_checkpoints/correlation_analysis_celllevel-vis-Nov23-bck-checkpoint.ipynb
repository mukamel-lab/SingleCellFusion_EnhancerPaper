```python
input_enh_gene_table = '/cndd2/fangming/projects/scf_enhancers/results/200521_to_evals.tsv' 
output_figures = '/cndd2/fangming/projects/scf_enhancers/results/figures/{}_metacell_corr_res_{{}}.pdf'.format(today)
```

```python
# enhancer-gene linkage
ti = time.time()
enhancer_gene_to_eval = (pd.read_csv(input_enh_gene_table, sep='\t')
                             [['gene', 'ens', 'dist']]
                             .rename(columns={'ens': 'enh'})
                        )

print(enhancer_gene_to_eval.shape)
print(time.time()-ti)
enhancer_gene_to_eval.head()
```

```python
bins = np.linspace(-1,1,201)
fdr_threshold = 0.2
distance_threshold = 1e5
label = 'pairs (<100kb)'

col_orders = [
            'num_clst',
    
            'num_linked_pairs',
            'num_linked_genes',
            'num_linked_enhs',
    
            'num_correlated_pairs',
            'num_correlated_genes',
            'num_correlated_enhs',
    
            'num_total_pairs',
            'num_total_genes',
            'num_total_enhs',
]
```

```python
knn = 30 # only 
f_pattern = ("/cndd2/fangming/projects/scf_enhancers/results/"
             "corr_analysis_10x_cells_v3_snmcseq_gene_mop_10x_cells_v3_snmcseq_gene_ka30_knn{}_201120"
             "_cluster_n[0-9]*_corrs.pkl".format(knn)
            )
fnames = np.sort(glob.glob(f_pattern))
iterator_mc = [(int(re.findall("_cluster_n[0-9]+", fname)[0][len("_cluster_n"):]), 
             fname,
            ) for fname in fnames]
print(len(fnames), fnames[:5])

f_pattern = ("/cndd2/fangming/projects/scf_enhancers/results/"
             "corr_analysis_10x_cells_v3_snatac_gene_mop_10x_cells_v3_snatac_gene_ka30_knn{}_201120"
             "_cluster_n[0-9]*_corrs.pkl".format(knn)
            )
fnames = np.sort(glob.glob(f_pattern))
iterator_atac = [(int(re.findall("_cluster_n[0-9]+", fname)[0][len("_cluster_n"):]), 
             fname,
            ) for fname in fnames]
print(len(fnames), fnames[:5])
```

```python
res_mc = []
for num_clst, output_corr in tqdm.tqdm(iterator_mc):
#     print(num_clst)
    # try loading the list
    try:
        with open(output_corr, 'rb') as fh:
            to_correlate, corrs, corrs_shuffled, corrs_shuffled_cells = pickle.load(fh)
#         print(to_correlate.shape, corrs.shape, corrs_shuffled.shape, corrs_shuffled_cells.shape)
    except:
        continue
        
    res_1case = enhancer_gene_utils.get_significance_stats_old(
        enhancer_gene_to_eval[to_correlate],
        corrs, corrs_shuffled, corrs_shuffled_cells, 
        bins=bins,
        distance_threshold=distance_threshold, fdr_threshold=fdr_threshold,
        positive_side=False,
        return_cdf=False,
        r_min=-0.9,
        r_max=0,
    )
    
    # record
    res_1case['num_clst'] = num_clst
    res_mc.append(pd.Series(res_1case)[col_orders])
    
res_mc = pd.DataFrame(res_mc)[col_orders] 
res_mc_mean = res_mc.groupby(['num_clst']).mean().reset_index()[col_orders]
res_mc_std = res_mc.groupby(['num_clst']).std().reset_index()[col_orders]
print(res_mc.shape)
res_mc.head()
```

```python
res_atac = []
for num_clst, output_corr in tqdm.tqdm(iterator_atac):
#     print(num_clst)
    # try loading the list
    try:
        with open(output_corr, 'rb') as fh:
            to_correlate, corrs, corrs_shuffled, corrs_shuffled_cells = pickle.load(fh)
#         print(to_correlate.shape, corrs.shape, corrs_shuffled.shape, corrs_shuffled_cells.shape)
    except:
        continue
        
    res_1case = enhancer_gene_utils.get_significance_stats_old(
        enhancer_gene_to_eval[to_correlate],
        corrs, corrs_shuffled, corrs_shuffled_cells, 
        bins=bins,
        distance_threshold=distance_threshold, fdr_threshold=fdr_threshold,
        positive_side=True,
        return_cdf=False,
        r_min=-0.9,
        r_max=0,
    )
    
    # record
    res_1case['num_clst'] = num_clst
    res_atac.append(pd.Series(res_1case)[col_orders])
    
res_atac = pd.DataFrame(res_atac)[col_orders] 
res_atac_mean = res_atac.groupby(['num_clst']).mean().reset_index()[col_orders]
res_atac_std = res_atac.groupby(['num_clst']).std().reset_index()[col_orders]
print(res_atac.shape)
res_atac.head()
```

```python
plot_cols = [
    'num_linked_pairs',
    'num_linked_genes',
    'num_linked_enhs',
#     'num_correlated_pairs',
#     'num_correlated_genes',
#     'num_correlated_enhs',
    'num_total_pairs',
    'num_total_genes',
    'num_total_enhs',
]

# plot same ks

mc_baseline = 6094+1300
atac_baseline = 2700+1300

fig, axs = plt.subplots(2, 3, figsize=(4*3, 3*2))
fig.suptitle('mC/ATAC-RNA (10x_cells_v3) cell correlation analysis (k={}; {}; FDR={})'.format(knn, label, fdr_threshold), 
             fontsize=15, y=1)
axs = axs.flatten()
for i_ax, (ax, col) in enumerate(zip(axs, plot_cols)):
    ymax = 0
    for _table, _table_std, _type in zip([res_mc_mean, res_atac_mean], 
                                         [res_mc_std, res_atac_std],
                                         ['mC-RNA', 'ATAC-RNA'],
                                        ):
    
        _x = _table['num_clst']
        _y = _table[col]
        _y_std = _table_std[col] 
        ax.plot(_x, _y, '-o', markersize=5, label=_type)
        ax.fill_between(_x, _y-_y_std, _y+_y_std, alpha=0.5)

        ymax = np.max([ymax, np.max(_y)])
    
    if i_ax == 0:
        ax.axhline(mc_baseline, color='C0', linestyle='--', alpha=0.5, label='mC-RNA (38 clusters)')
        ax.axhline(atac_baseline, color='C1', linestyle='--', alpha=0.5, label='ATAC-RNA (38 clusters)')
        
    ax.set_xscale('log')
    ax.set_xlim([5, 20000])
    ax.set_xticks([10, 100, 1000, 10000])
    ax.set_ylim(ymin=0, ymax=1.1*ymax)
    ax.set_ylabel(col.replace('_', ' '))

    ax.xaxis.set_major_formatter(mtick.EngFormatter())
    ax.yaxis.set_major_formatter(mtick.EngFormatter())
    ax.set_xlabel('Num clusters')
        
# axs[-1].axis('off') 
handles, labels = snmcseq_utils.combine_legends(axs)
handles, labels = snmcseq_utils.dedup_legends(handles, labels)
axs[2].legend(handles, labels, bbox_to_anchor=(1,1), loc='upper left')
fig.subplots_adjust(wspace=0.5, hspace=0.5)
snmcseq_utils.savefig(fig, output_figures.format('p1_k{}'.format(knn)))
plt.show()
```

```python
bins = np.linspace(-1,1,201)
fdr_threshold = 0.2
distance_threshold = 1e5
label = 'pairs (<100kb)'

col_orders = [
            'num_clst',
    
            'num_linked_pairs',
            'num_linked_genes',
            'num_linked_enhs',
            'id_linked_pairs',
    
            'num_correlated_pairs',
            'num_correlated_genes',
            'num_correlated_enhs',
            'id_correlated_pairs',
    
            'num_total_pairs',
            'num_total_genes',
            'num_total_enhs',
            'id_total_pairs',
]
```

```python
knn = 30 # only 
f_pattern = ("/cndd2/fangming/projects/scf_enhancers/results/"
             "corr_analysis_10x_cells_v3_snmcseq_gene_mop_10x_cells_v3_snmcseq_gene_ka30_knn{}_201120"
             "_cluster_n[0-9]*_corrs.pkl".format(knn)
            )
fnames = np.sort(glob.glob(f_pattern))
iterator_mc = [(int(re.findall("_cluster_n[0-9]+", fname)[0][len("_cluster_n"):]), 
             fname,
            ) for fname in fnames]

f_pattern = ("/cndd2/fangming/projects/scf_enhancers/results/"
             "corr_analysis_10x_cells_v3_snatac_gene_mop_10x_cells_v3_snatac_gene_ka30_knn{}_201120"
             "_cluster_n[0-9]*_corrs.pkl".format(knn)
            )
fnames = np.sort(glob.glob(f_pattern))
iterator_atac = [(int(re.findall("_cluster_n[0-9]+", fname)[0][len("_cluster_n"):]), 
             fname,
            ) for fname in fnames]

# match file names if both exists
iterator_both = pd.merge(pd.DataFrame(iterator_mc, columns=['num_clst', 'fname_mc']), 
                         pd.DataFrame(iterator_atac, columns=['num_clst', 'fname_atac']), 
                         on='num_clst', how='inner').astype({'num_clst': int}).sort_values('num_clst')
print(iterator_both.shape)
iterator_both.head()
```

```python
positive_sides = [False, True]

res = []
# res_mean = [[], []]
# res_std = [[], []]

for idx, row in tqdm.tqdm(iterator_both.iterrows()):
    num_clst, fname_mc, fname_atac = (row['num_clst'], 
                                      row['fname_mc'],
                                      row['fname_atac'],
                                     )
    res_2cases = []
    for i, (label, fname, positive_side) in enumerate(zip(
                        ['mc', 'atac'],
                        [fname_mc, fname_atac], 
                        [False, True],
                        )):
        try:
            with open(fname, 'rb') as fh:
                to_correlate, corrs, corrs_shuffled, corrs_shuffled_cells = pickle.load(fh)
        except:
            continue
            
        pairs = enhancer_gene_to_eval[to_correlate]
        label_cond = pairs['dist'].values < distance_threshold
            
        res_1case = enhancer_gene_utils.get_significance_stats_old(
            pairs,
            corrs, corrs_shuffled, corrs_shuffled_cells, 
            bins=bins,
            distance_threshold=distance_threshold, fdr_threshold=fdr_threshold,
            positive_side=positive_side,
            return_cdf=False,
            r_min=-0.9,
            r_max=0,
        )

        # record
        res_1case['num_clst'] = num_clst
        res_1case['id_total_pairs'] = pairs[label_cond].index.values 
        
        if isinstance(res_1case['linked_table'], pd.DataFrame):
            res_1case['id_linked_pairs'] = res_1case['linked_table'].index.values 
        else:
            res_1case['id_linked_pairs'] = np.array([])  
            
        if isinstance(res_1case['correlated_table'], pd.DataFrame):
            res_1case['id_correlated_pairs'] = res_1case['correlated_table'].index.values 
        else:
            res_1case['id_correlated_pairs'] = np.array([]) 
            
        res_1case = pd.Series(res_1case)[col_orders]
        res_1case.index = res_1case.index + '_{}'.format(label)
        res_2cases.append(res_1case)
        
    # coommon
    res_2cases = pd.concat(res_2cases)
    
    # linked both
    common_pair_ids = np.intersect1d(res_2cases['id_linked_pairs_mc'],
                                     res_2cases['id_linked_pairs_atac'],
                                    )
    common_gene_ids = enhancer_gene_to_eval.loc[common_pair_ids, 'gene'].unique()
    common_enh_ids = enhancer_gene_to_eval.loc[common_pair_ids, 'enh'].unique()
    
    res_2cases['num_linked_pairs_both'] = len(common_pair_ids)
    res_2cases['num_linked_genes_both'] = len(common_gene_ids)
    res_2cases['num_linked_enhs_both'] = len(common_enh_ids)
    
    # total both
    common_pair_ids = np.intersect1d(res_2cases['id_total_pairs_mc'],
                                     res_2cases['id_total_pairs_atac'],
                                    )
    common_gene_ids = enhancer_gene_to_eval.loc[common_pair_ids, 'gene'].unique()
    common_enh_ids = enhancer_gene_to_eval.loc[common_pair_ids, 'enh'].unique()
    
    res_2cases['num_total_pairs_both'] = len(common_pair_ids)
    res_2cases['num_total_genes_both'] = len(common_gene_ids)
    res_2cases['num_total_enhs_both'] = len(common_enh_ids)
    
    res.append(pd.Series(res_2cases))
res = pd.DataFrame(res)
print(res.shape)
res.head()

res_mean = res.groupby(['num_clst_mc']).mean().reset_index()
res_std = res.groupby(['num_clst_mc']).std().reset_index()
print(res_mean.shape, res_std.shape)
```

```python
plot_cols = [
    'num_linked_pairs',
    'num_linked_genes',
    'num_linked_enhs',
    
    'num_total_pairs',
    'num_total_genes',
    'num_total_enhs',
]

# plot same ks

_table = res_mean
_table_std = res_std

mc_baseline = 6094+1300
atac_baseline = 2700+1300
both_baseline = 1300

fig, axs = plt.subplots(2, 3, figsize=(4*3, 3*2))
fig.suptitle('mC/ATAC-RNA (10x_cells_v3) cell correlation analysis (k={}; {}; FDR={})'.format(knn, label, fdr_threshold), 
             fontsize=15, y=1)
axs = axs.flatten()
for i_ax, (ax, col) in enumerate(zip(axs, plot_cols)):
    ymax = 0
    for _suffix, _type in zip(['_mc', '_atac', '_both'], 
                              ['mC-RNA', 'ATAC-RNA', 'both'],
                             ):
        _x = _table['num_clst_mc']
        _y = _table[col+_suffix]
        _y_std = _table_std[col+_suffix] 
        ax.plot(_x, _y, '-o', markersize=5, label=_type)
        ax.fill_between(_x, _y-_y_std, _y+_y_std, alpha=0.5)
        ymax = np.max([ymax, np.max(_y)])
    
    if i_ax == 0:
        ax.axhline(mc_baseline, color='C0', linestyle='--', alpha=0.5, label='mC-RNA (38 clusters)')
        ax.axhline(atac_baseline, color='C1', linestyle='--', alpha=0.5, label='ATAC-RNA (38 clusters)')
        ax.axhline(both_baseline, color='C2', linestyle='--', alpha=0.5, label='both (38 clusters)')
        
    ax.set_xscale('log')
    ax.set_xlim([5, 20000])
    ax.set_xticks([10, 100, 1000, 10000])
    ax.set_ylim(ymin=0, ymax=1.1*ymax)
    ax.set_ylabel(col.replace('_', ' '))

    ax.xaxis.set_major_formatter(mtick.EngFormatter())
    ax.yaxis.set_major_formatter(mtick.EngFormatter())
    ax.set_xlabel('Num clusters')
        
# axs[-1].axis('off') 
handles, labels = snmcseq_utils.combine_legends(axs)
handles, labels = snmcseq_utils.dedup_legends(handles, labels)
axs[2].legend(handles, labels, bbox_to_anchor=(1,1), loc='upper left')
fig.subplots_adjust(wspace=0.5, hspace=0.5)
snmcseq_utils.savefig(fig, output_figures.format('p1_k{}'.format(knn)))
plt.show()
```

```python
plot_cols = [
    'num_linked_pairs',
    'num_linked_genes',
    'num_linked_enhs',
    
    'num_total_pairs',
    'num_total_genes',
    'num_total_enhs',
]

# plot same ks

_table = res_mean
_table_std = res_std

mc_baseline = 6094+1300
atac_baseline = 2700+1300
both_baseline = 1300

fig, axs = plt.subplots(2, 3, figsize=(4*3, 3*2))
fig.suptitle('mC/ATAC-RNA (10x_cells_v3) cell correlation analysis (k={}; {}; FDR={})'.format(knn, label, fdr_threshold), 
             fontsize=15, y=1)
axs = axs.flatten()
for i_ax, (ax, col) in enumerate(zip(axs, plot_cols)):
    ymax = 0
    for _suffix, _type in zip(['_mc', '_atac', '_both'], 
                              ['mC-RNA', 'ATAC-RNA', 'both'],
                             ):
        _x = _table['num_clst_mc']
        _y = _table[col+_suffix]
        _y_std = _table_std[col+_suffix] 
        ax.plot(_x, _y, '-o', markersize=5, label=_type)
        ax.fill_between(_x, _y-_y_std, _y+_y_std, alpha=0.5)
        ymax = np.max([ymax, np.max(_y)])
    
    if i_ax == 0:
        ax.axhline(mc_baseline, color='C0', linestyle='--', alpha=0.5, label='mC-RNA (38 clusters)')
        ax.axhline(atac_baseline, color='C1', linestyle='--', alpha=0.5, label='ATAC-RNA (38 clusters)')
        ax.axhline(both_baseline, color='C2', linestyle='--', alpha=0.5, label='both (38 clusters)')
        
    ax.set_xscale('log')
    ax.set_xlim([5, 20000])
    ax.set_xticks([10, 100, 1000, 10000])
    ax.set_ylim(ymin=0, ymax=1.1*ymax)
    ax.set_ylabel(col.replace('_', ' '))

    ax.xaxis.set_major_formatter(mtick.EngFormatter())
    ax.yaxis.set_major_formatter(mtick.EngFormatter())
    ax.set_xlabel('Num clusters')
        
# axs[-1].axis('off') 
handles, labels = snmcseq_utils.combine_legends(axs)
handles, labels = snmcseq_utils.dedup_legends(handles, labels)
axs[2].legend(handles, labels, bbox_to_anchor=(1,1), loc='upper left')
fig.subplots_adjust(wspace=0.5, hspace=0.5)
snmcseq_utils.savefig(fig, output_figures.format('p1_k{}'.format(knn)))
plt.show()
```
