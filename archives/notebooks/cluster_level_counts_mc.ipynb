```python
f_mc = '/cndd/fangming/CEMBA/data/MOp_all/data_freeze_neurons/snmcseq_gene_mCH_raw.npz'
f_c = '/cndd/fangming/CEMBA/data/MOp_all/data_freeze_neurons/snmcseq_gene_CH_raw.npz' 
f_cell = '/cndd/fangming/CEMBA/data/MOp_all/data_freeze_neurons/snmcseq_gene_raw.cell'
f_gene = '/cndd/fangming/CEMBA/data/MOp_all/data_freeze_neurons/snmcseq_gene_raw.gene'

gc_matrix_mch = snmcseq_utils.load_gc_matrix_methylation(f_gene, f_cell, f_mc, f_c)
```

```python
# get mCH for each cluster
_x = gc_matrix_mch 
intg_clst = df_clst
_xtmp_data_c = _x.data['c'].tocsc()
_xtmp_data_mc = _x.data['mc'].tocsc()

for _round in ['round2', 'round3']:
    cluster_level = 'joint_cluster_{}'.format(_round)
    clusters = np.sort(df_clst[cluster_level].unique())
    
    gene_mch_c_clsts = pd.DataFrame(index=_x.gene, columns=clusters)
    gene_mch_mc_clsts = pd.DataFrame(index=_x.gene, columns=clusters)
    global_mean_mch = {}
    global_mean_mcg = {}
    for clst, df_sub in intg_clst[intg_clst['modality']=='snmcseq_gene'].groupby(cluster_level):
        cells = df_sub['sample'].values 
        cells = np.intersect1d(cells, _x.cell)
#         global_mean_mch[clst] = df_meta.loc[cells, 'CH_Rate'].mean()
#         global_mean_mcg[clst] = df_meta.loc[cells, 'CG_Rate'].mean()
        cells_idx = snmcseq_utils.get_index_from_array(_x.cell, cells)

        gene_mch_c_clsts[clst] = _xtmp_data_c[:, cells_idx].sum(axis=1) 
        gene_mch_mc_clsts[clst] = _xtmp_data_mc[:, cells_idx].sum(axis=1) 
#     global_mean_mch = pd.Series(global_mean_mch)
#     global_mean_mcg = pd.Series(global_mean_mcg)

    f = '/cndd2/fangming/miniatlas/cluster_level/gene_counts_mch_ch_{}.tsv'.format(_round)
    gene_mch_c_clsts.to_csv(f, na_rep='NA', sep="\t", index=True, header=True)
    f = '/cndd2/fangming/miniatlas/cluster_level/gene_counts_mch_mch_{}.tsv'.format(_round)
    gene_mch_mc_clsts.to_csv(f, na_rep='NA', sep="\t", index=True, header=True)
```

```python
# global methylation level
intg_clst = df_clst
all_clusters = np.hstack([
    np.sort(df_clst['joint_cluster_round2'].unique()),
    np.sort(df_clst['joint_cluster_round3'].unique()),
])
global_mean_mch = pd.Series(index=all_clusters)
global_mean_mcg = pd.Series(index=all_clusters)
for _round in ['round2', 'round3']:
    cluster_level = 'joint_cluster_{}'.format(_round)
    for clst, df_sub in intg_clst[intg_clst['modality']=='snmcseq_gene'].groupby(cluster_level):
        cells = df_sub['sample'].values 
        cells = np.intersect1d(cells, _x.cell)
        global_mean_mch.loc[clst] = df_meta.loc[cells, 'CH_Rate'].mean()
        global_mean_mcg.loc[clst] = df_meta.loc[cells, 'CG_Rate'].mean()
```

```python
# global_mc = pd.DataFrame()
global_mc = pd.merge(global_mean_mch.to_frame('mch'), 
                     global_mean_mcg.to_frame('mcg'), 
                     left_index=True, right_index=True,
                    )
fout = '/cndd2/fangming/miniatlas/cluster_level/global_mc.tsv'
global_mc.to_csv(fout, sep='\t', na_rep='NA', header=True, index=True)
```

```python
for _round in rounds:
    # mch specs
    global_mean_mcc = global_mean_mch
    c_file = '/cndd2/fangming/miniatlas/cluster_level/gene_counts_mch_ch_{}.tsv'.format(_round)
    mc_file = '/cndd2/fangming/miniatlas/cluster_level/gene_counts_mch_mch_{}.tsv'.format(_round)
    mcc_file_out = '/cndd2/fangming/miniatlas/cluster_level/gene_features_mch_{}.tsv'.format(_round)
    # end of mch specs

    gene_c_clsts = pd.read_csv(c_file, sep="\t", index_col=0)
    gene_mc_clsts = pd.read_csv(mc_file, sep="\t", index_col=0)
    
    base_call_cutoff = 100
    # get mcc
    gene_mcc_clsts = snmcseq_utils.get_mcc_lite_v3(gene_c_clsts, gene_mc_clsts, base_call_cutoff=base_call_cutoff) # some entries will be nan
    gene_mcc_clsts = gene_mcc_clsts.divide(global_mean_mcc.loc[gene_mcc_clsts.columns.values], axis=1)
    gene_mcc_clsts.to_csv(mcc_file_out, sep='\t', na_rep='NA', header=True, index=True)
    print(gene_mcc_clsts.shape)
    gene_mcc_clsts.head()
```
